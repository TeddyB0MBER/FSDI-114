```python
class DNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
class DoublyLinkedList:
    def __init__(self):
        # self.head is first node in the list
        self.head = None
        #this function adds a new node if head is empty
    def append(self, element):
        new_node = DNode(element)
        if not self.head: 
            #this means the list is empty 
            self.head = new_node
        else: 
            current  = self.head 
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current 
            
    def __str__(self):
        out_str = "["
        if self.head:
            #  %s is a way to format a string 
            out_str += "%s" % self.head.data
            current = self.head
            while current:
                current = current.next
                if current: 
                    out_str += ", %s" % current.data
        out_str += "]"
        return out_str
    
    def remove(self,value):
        found = False
        if self.head: 
            current = self.head 
            while current.data != value and current.next:
                current = current.next 
            if current.data == value:
                prev_node = current.prev
                next_node = current.next 
                prev_node.next = next_node
                next_node.prev = prev_node
                found = True
        if not found:
            raise ValueError("Value not present in list")
            
    def insert(self, index, value):
        new_node = DNode(value)
        if not self.head:
            self.head = new_node    
        else: 
            if index == 0:
                current_head = self.head
                new_node.next = current_head
                current_head.prev = new_node
                self.head = new_node
            else: 
                current = self.head
                counter  = 0
                while current.next and counter != index:
                    current = current.next
                    counter += 1 
                if counter == index: 
                    prev_node = current.prev 
                    prev_node.next = new_node
                    new_node.prev = prev_node
                    new_node.next = current
                    current.prev = new_node
                else:
                    current.next = new_node
                    new_node.prev = current
                    
            
```

```python
dll = DoublyLinkedList()

for i in range(1,11):
    dll.append(i)
print("Append numbers 1 to 10:")
print("_"*30) 
print(dll)

print("remove number 5 from list:")
print("-"*30)
dll.remove(5)
print(dll)


dll2 = DoublyLinkedList()

print("inseert a new node at head of an empty DLL")
print("-"*30)
dll2.insert(0,1
           )
print("insert a new node before an existing head")
print("-" *30)
# With 0 set as index:
dll2.insert(0, 0)
print(dll2)

print("Insert a new node anywhere in the middle")
print("-" *30)
dll2.insert(1, 0.5)

```
